# 指标速览
这张表用于在同一页面每日抓取数据，观察异常指标并追踪

## Macroeconomics

```python
import pandas_datareader.data as web
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import datetime

# Set style
plt.style.use('ggplot')
start_date = datetime.datetime(2020, 1, 1)
end_date = datetime.datetime.now()

# 1. FRED Economic Indicators
print("Fetching FRED data...")
fred_tickers = ['GDP', 'CPIAUCSL', 'UNRATE', 'FEDFUNDS']
try:
    fred_data = web.DataReader(fred_tickers, 'fred', start_date, end_date)
    
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
    fig.suptitle('Macroeconomic Indicators (FRED)', fontsize=16)
    
    fred_data['GDP'].plot(ax=axes[0,0], title='US GDP')
    axes[0,0].set_ylabel('Billions of Dollars')
    
    fred_data['CPIAUCSL'].plot(ax=axes[0,1], title='US CPI')
    axes[0,1].set_ylabel('Index 1982-1984=100')
    
    fred_data['UNRATE'].plot(ax=axes[1,0], title='Unemployment Rate')
    axes[1,0].set_ylabel('Percent')
    
    fred_data['FEDFUNDS'].plot(ax=axes[1,1], title='Federal Funds Rate')
    axes[1,1].set_ylabel('Percent')
    
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Error fetching FRED data: {e}")

# 2. Market Indicators
print("Fetching Market data...")
market_tickers = ['^GSPC', '^TNX', '^VIX', 'DX-Y.NYB']
try:
    # auto_adjust=True is default now, so we get 'Close' which is adjusted
    market_data = yf.download(market_tickers, start=start_date, end=end_date)['Close']
    
    fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 10))
    fig.suptitle('Market Indicators', fontsize=16)
    
    market_data['^GSPC'].plot(ax=axes[0,0], title='S&P 500')
    axes[0,0].set_ylabel('Price')
    
    market_data['^TNX'].plot(ax=axes[0,1], title='10Y Treasury Yield')
    axes[0,1].set_ylabel('Yield (%)')
    
    market_data['^VIX'].plot(ax=axes[1,0], title='VIX Index')
    axes[1,0].set_ylabel('Index')
    
    market_data['DX-Y.NYB'].plot(ax=axes[1,1], title='US Dollar Index')
    axes[1,1].set_ylabel('Index')
    
    plt.tight_layout()
    plt.show()
except Exception as e:
    print(f"Error fetching Market data: {e}")
```
